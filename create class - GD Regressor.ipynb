```python
class GDRegressor:

    def __init__(self,learning_rate,epochs):

        self.m = 29.19
        self.b = -120
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self,X,y):
        # calculate the b using Gradient Descent
        for i in range(self.epochs):
            loss_slope = -2 * np.sum(y - self.m*X.ravel() - self.b)
            self.b = self.b - (self.lr * loss_slope)
        print(self.b)
```

```python
gd = GDRegressor(0.1,10)
```

```python
gd.fit(X,y) # Yen muze batayega ki 10 epoch run krane ke baad mera slop kitana aayega
```
